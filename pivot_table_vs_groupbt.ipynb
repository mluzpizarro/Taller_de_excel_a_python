## **pivot_table vs groupby**

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                          "bar", "bar", "bar", "bar"],
                    "B": ["one", "one", "one", "two", "two",
                          "one", "one", "two", "two"],
                    "C": ["small", "large", "large", "small",
                          "small", "large", "small", "small",
                          "large"],
                    "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                    "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
df
```

* 1 sola variable en filas

```python
df[['A','B','D']].groupby('A').mean()
```

```python
pd.pivot_table(df, values='D', index=['A'], aggfunc=np.mean)
```

* 2 variables en filas

```python
df[['A','B','D']].groupby(['A','B']).mean()
```

```python
pd.pivot_table(df, values='D', index=['A','B'], aggfunc=np.mean)
```

* 3 variables en filas

```python
df[['A','B','C','D']].groupby(['A','B','C']).mean()
```

```python
pd.pivot_table(df, values='D', index=['A','B','C'], aggfunc=np.mean)
```

* 2 variable en fila y una en columna

```python
df[['A','B','C','D']].groupby(['A','B','C']).mean().unstack('A').fillna(0)
```

```python
pd.pivot_table(df, values='D', index=['B','C'], columns='A', aggfunc=np.mean, fill_value=0)
```

* 2 variable en fila y 2 en columna

```python
df[['A','B','C','D']].groupby(['A','B','C']).mean().unstack('A').unstack('B').fillna(0)
```

```python
pd.pivot_table(df, values='D', index=['C'], columns=['A','B'], aggfunc=np.mean, fill_value=0)
```

```python
pd.pivot_table(df, values='D', index=['A', 'B'],
                     columns=['C'], aggfunc=np.sum)
```

```python
pd.pivot_table(df, values='D', index=['B'],
                     columns=['C','A'], aggfunc=np.sum, fill_value=0)
```

**Unstack**  
Lo mismo pero con unstack, esta funcion es equivalente a poner una columna de la tabla como col de una tabla dinamica

```python
pd.pivot_table(df, values='D', index=['A', 'B'],
                     columns=['C'], aggfunc=np.sum, fill_value=0).unstack('A')
```

```python
pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                     aggfunc={'D': np.mean,
                              'E': np.mean},margins=True)
```

```python
pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                     aggfunc={'D': np.mean,
                              'E': [min, max, np.mean]})
```
